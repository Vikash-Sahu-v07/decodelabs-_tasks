```python
import pandas as pd
import sqlite3
```

```python
# 1. Load your cleaned dataset
df = pd.read_excel("Cleaned_Dataset_Data_Analytics.xlsx")

# 2. Create an in-memory SQL database connection
conn = sqlite3.connect(':memory:')

# 3. Dump the Excel data into a SQL table named 'orders'
df.to_sql('orders', conn, index=False, if_exists='replace')

print("SQL Database Engine Ready! Your 'orders' table is populated.")
```

```python
# Type this in your next cell to see your SQL table data!
pd.read_sql_query("SELECT * FROM orders LIMIT 5;", conn)
```

```python
#Cell 1: Query 1 — Total Revenue by Product Category

# %% [markdown]
# ### Query 1: Realized Total Revenue by Product Category (Excludes Cancellations)

# %%
query_1 = """
SELECT 
    Product,
    COUNT(OrderID) AS Total_Orders,
    SUM(TotalPrice) AS Total_Revenue
FROM 
    orders
WHERE 
    OrderStatus != 'Cancelled'
GROUP BY 
    Product
ORDER BY 
    Total_Revenue DESC;
"""

# Run the query and view the results table
df_result_1 = pd.read_sql_query(query_1, conn)
df_result_1
```

```python
# %% [markdown]
# ### Query 2: Payment Methods with Average Order Value Over $1,000

# %%
query_2 = """
SELECT 
    PaymentMethod,
    COUNT(OrderID) AS Order_Count,
    AVG(TotalPrice) AS Average_Order_Value
FROM 
    orders
GROUP BY 
    PaymentMethod
HAVING 
    AVG(TotalPrice) > 1000.00
ORDER BY 
    Average_Order_Value DESC;
"""

df_result_2 = pd.read_sql_query(query_2, conn)
df_result_2
```

```python
# %% [markdown]
# ### Query 3: Lost Revenue Matrix From Cancelled Orders by Marketing Channel

# %%
query_3 = """
SELECT 
    ReferralSource,
    COUNT(OrderID) AS Cancelled_Orders_Count,
    SUM(TotalPrice) AS Lost_Revenue
FROM 
    orders
WHERE 
    OrderStatus = 'Cancelled'
GROUP BY 
    ReferralSource
ORDER BY 
    Lost_Revenue DESC;
"""

df_result_3 = pd.read_sql_query(query_3, conn)
df_result_3
```
